```python
# Given a dataset containing details about California housing, such as average income, house age, number of rooms, population, and other features, build a machine linear regression model that can predict the expected price of a house in California.

# Target : Median house value in thousands of dollars.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Step 2: Load the California Housing dataset into housing which is variable/object
housing = fetch_california_housing()   
```

```python
# Step 3: Create X and Y
X = housing.data    # Features
Y = housing.target  # Target Variable(House Price)
```

```python
# Check the shape
# 20640: The number of samples (houses) in the dataset.
# 8: The number of features (input variables) used to describe each house.

print(f"Shape of X : {X.shape}")
print(f"Shape of Y : {Y.shape}")
```

```python
# Step 4 : Normalize the feature

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Step 5: Split data into training and testing

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Step 6: Create and train Linear Regression model

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
# Step 7: Make Predictions on test data

Y_pred = model.predict(X_test)  # Taking output in Y_pred from the input(X_test: which is new unseen example for our model)
```

```python
# Step 8: Evaluate the model, Calculating its MSE and R^2 to do sorted

from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)               # Can't Swap (Y_test and Y_pred)...this order shall be maintained
```

```python
print(f"Mean Squared Error : {mse}")
print(f"R-Squared Value : {r2}")
```

```python
# Visualizing the model using matplotlib

# 1. Scatter plot of v

plt.figure(figsize=(10,6))    # size of figure which will be width as 10 and height as 6
plt.scatter(Y_test, Y_pred, color = 'blue')
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], color='black', linewidth=2)
plt.title("Actual Vs Predicted House Prices")
plt.xlabel("Actual House Prices")
plt.ylabel("Predicted House Prices")
plt.grid(True)
plt.show()
```

```python
# Residual Plot

residuals = Y_test - Y_pred

plt.figure(figsize = (10,6))
plt.scatter(Y_pred, residuals, color = 'red')
plt.axhline(y=0, color = 'black', linestyle='--')
plt.title('Residual Plot')
plt.xlabel('Predicted House Prices')
plt.ylabel("Residuals")
plt.grid(True)
plt.show()
```
