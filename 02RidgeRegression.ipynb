```python
# Step 1: Import required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Step 2 : Fetch the data from dataset into housing

housing = fetch_california_housing()
```

```python
# Step 3 :
# Create X and Y

# X: Input Features (independent variables) → used to predict.
# Y: Target Output

X = housing.data
Y = housing.target
```

```python
# Step 4: Normalize (Scale) the Features
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Step 5 : Split data into train and test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Step 6: Create and Train the Ridge Regression model

ridge_model = Ridge(alpha=1)  # alpha controls the regularization strength
ridge_model.fit(X_train, Y_train)
```

```python
# Step 7: Make Predictions on Test Data
Y_pred = ridge_model.predict(X_test)
```

```python
# Step 8: Evaluate the Model (Calculate MSE and R²)

mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f"Mean Squared Error is : {mse}")
print(f"R Square value is : {r2}")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(Y_test, Y_pred, color = 'blue')
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], color = 'black', linewidth = 2)
plt.title("Actual Vs Predicted House Prices")
plt.xlabel("Actual House Prices")
plt.ylabel("Predicted House Prices")
plt.grid(True)
plt.show()
```

```python
# Residuals are the differences between the actual values and the predicted values.
# They represent how much the model's predictions deviate from the real values.

residuals = Y_test - Y_pred
plt.figure(figsize=(10,6))
plt.scatter(Y_pred, residuals, color = 'red')
plt.axhline(y=0, color = 'black', linestyle = '-.')
plt.title("Residual Plot")
plt.xlabel("Predicted House Prices")
plt.ylabel("Residuals")
plt.grid(True)
plt.show()
```
